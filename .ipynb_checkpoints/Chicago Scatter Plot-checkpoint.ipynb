```python
%matplotlib notebook
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as st
from scipy.stats import linregress
from sklearn import datasets
csv=pd.read_csv("Data\\Energy_Usage_2010.csv")
Summary=csv
Summary=Summary.dropna(how='any')
residential_summary = Summary.loc[(Summary['BUILDING TYPE'] == 'Residential') & (Summary['BUILDING_SUBTYPE'] == 'Single Family')]
Summary.columns
Yearly_Summary=residential_summary['KWH TOTAL SQFT']
x_axis=residential_summary['TOTAL POPULATION']
y_axis=Yearly_Summary
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_axis, y_axis)
regress_values = x_axis * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.plot(x_axis,regress_values,"r-")
plt.annotate(line_eq,(9,10),fontsize=10,color="red")
plt.scatter(x_axis,y_axis)
plt.xlabel('Residential Population')
plt.ylabel('Electricity Consumed(KWH)')
print(f"The r-squared is: {rvalue**2}")
plt.show()

```
